```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# k: 100, N_points = 10k
sequential = 4713450
threads_map = np.array([4929921,2634814,1353940,825423,497499,340730,230096,155614,153191])
threads_pool = np.array([5228737,2636180,1376279,827818,538796,358711,360791,391634,452312])
ff_parallel_reduce = np.array([5220850,2651574,1348746,731048,381185,228666,214618,246259,406376])
ff_farm = np.array([5519480,2840748,1672978,1065591,408902,251250,197531,242681,393676])
```

```python
speedup_threads_map = sequential/threads_map
speedup_threads_pool = sequential/threads_pool
speedup_ff_parallel_reduce = sequential/ff_parallel_reduce 
speedup_ff_farm = sequential/ff_farm
idea = [1,2,4,8,16,32,64,128,256]
plt.figure(figsize=(10,6))
plt.xlabel('nWorkers')
plt.ylabel('Tseq/Tpar(n)')
plt.title('SpeedUp')
plt.xticks([i for i in range(9)],[1,2,4,8,16,32,64,128,256])
plt.axis([0, 8, 0, 256])
plt.plot(speedup_threads_map)
plt.plot(speedup_threads_pool)
plt.plot(speedup_ff_parallel_reduce)
plt.plot(speedup_ff_farm)
plt.plot(idea)
plt.legend(('threads_map', 'threads_pool', 'ff_parallel_reduce', 'ff_farm', 'idea'), loc='best')
plt.show()
```

```python
scalab_threads_map = threads_map[0]/threads_map
scalab_threads_pool = threads_pool[0]/threads_pool
scalab_ff_parallel_reduce = ff_parallel_reduce[0]/ff_parallel_reduce
scalab_ff_farm = ff_farm[0]/ff_farm
idea = [1,2,4,8,16,32,64,128,256]
plt.figure(figsize=(10,6))
plt.xlabel('nWorkers')
plt.ylabel('Tpar(1)/Tpar(n)')
plt.title('Scalability')
plt.xticks([i for i in range(9)],[1,2,4,8,16,32,64,128,256])
plt.axis([0, 8, 0, 256])
plt.plot(scalab_threads_map)
plt.plot(scalab_threads_pool)
plt.plot(scalab_ff_parallel_reduce)
plt.plot(scalab_ff_farm)
plt.plot(idea)
plt.legend(('threads_map', 'threads_pool', 'ff_parallel_reduce', 'ff_farm', 'idea'), loc='best')
plt.show()
```

```python
n = [1,2,4,8,16,32,64,128,256]
effi_threads_map = sequential/(n*threads_map)
effi_threads_pool = sequential/(n*threads_pool)
effi_ff_parallel_reduce = sequential/(n*ff_parallel_reduce)
effi_ff_farm = sequential/(n*ff_farm)
idea = np.ones(9)
plt.figure(figsize=(8,6))
plt.xlabel('nWorkers')
plt.ylabel('Tseq/(n*Tpar(n))')
plt.title('Efficiency')
plt.xticks([i for i in range(9)],[1,2,4,8,16,32,64,128,256])
plt.axis([0, 8, 0, 1.2])
plt.plot(effi_threads_map)
plt.plot(effi_threads_pool)
plt.plot(effi_ff_parallel_reduce)
plt.plot(effi_ff_farm)
plt.plot(idea)
plt.legend(('threads_map', 'threads_pool', 'ff_parallel_reduce', 'ff_farm', 'idea'), loc='best')
plt.show()
```
